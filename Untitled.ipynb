```python
def softmax(weights, beta=1):
    sum_exp_weights = sum([np.exp(beta*w) for w in weights])
    probabilities = np.array([np.exp(beta*w) for w in weights]) / sum_exp_weights
    return probabilities
```

```python
def resample_positions_softmax(weights, positions, beta=1):
    probabilities = softmax(weights, beta)
    pos_filter =  np.random.choice(list(range(len(positions))), len(positions), p=probabilities)
    return np.array(positions)[np.array(pos_filter)]
```

```python
def diffusion_interaction(U, grad_U, process, total_iter, verbose=False, domain_enforcer=None):
    p_start = process["start"]
    p_gamma = process["gamma"]
    p_temperature = process["temperature"]
    p_num_particles = len(p_start)
    p_epsilon = process["epsilon"]
    p_weight_func = process["weight_function"]
    p_weight_decay = process["weight_decay"]
    p_force_func = process["interactive_force_function"]
    p_interaction_weight = process["interaction_weight"]
    p_weight_probability_function = process["weight_to_probability"]

    dim = len(p_start)

    # init num_particles
    all_paths = [[np.array(p)] for p in p_start]
    p_weights = np.zeros(len(p_start))

    for t in range(total_iter):
        # force matrix 
        force_matrix = np.array([[p_force_func(all_paths[i][-1], all_paths[j][-1]) 
                         for j in range(len(all_paths))] for i in range(len(all_paths))])
                
        softmax_weights = p_weight_probability_function(p_weights)
        
        
        for i in range(p_num_particles):
            # find force on particle 
            p_force = np.sum((force_matrix[i].T * np.concatenate([softmax_weights[i:], softmax_weights[:i]])).T, axis=0)
                 
            # --- diffusion step ---
            x_curr = all_paths[i][-1]

            x_next = x_curr + p_gamma(t) * (
                -grad_U(x_curr)) + p_interaction_weight*p_force + p_temperature(t) * np.array(
                [[np.random.normal()] for _ in range(x_curr.shape[0])]).reshape(x_curr.shape)

            if domain_enforcer is not None:
                x_next, went_outside_domain = domain_enforcer(x_next)

            # ----

            all_paths[i].append(x_next)

            # weight update
            p_weights[i] = p_weight_func(U, grad_U, x_next) + p_weight_decay * p_weights[i]


    return np.array(all_paths)
```
