```python
import libsimulation
import pandas as pd
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt
import pickle
import time
```

```python
data=pd.read_pickle("test_pickle.pkl") #pd.read_csv("test_csv.csv")
```

```python
# =============================================================================
# Player Stat Team Organizer
# =============================================================================

sorted_data = data.sort_values(by=["name","dateTime"])
og_data = data.sort_values(by=["name","dateTime"])
gamma = 0.3


#Initialize
injury_status = 0
injury_index = 0


for i in range(0,sorted_data.shape[0]):
    
    #Initialize (Save even if injury == True)
    if i == 0:
        start_index = i 
        m=sorted_data.iloc[0,8]
        if m != 0: #If minute is not zero, else leave the values at zero
            sorted_data.iloc[0,8]=sorted_data.iloc[0,8]  #Minutes
            sorted_data.iloc[0,9]=sorted_data.iloc[0,9]/m #Points
            sorted_data.iloc[0,11]=sorted_data.iloc[0,11]/m #Rebounds
            sorted_data.iloc[0,12]=sorted_data.iloc[0,12]/m #Steals
            sorted_data.iloc[0,5]=sorted_data.iloc[0,5]/m #Block
    
    #For one player
    if sorted_data.iloc[i-1,1] == sorted_data.iloc[i,1]:       
        if sorted_data.iloc[i,13] == False: #Not injuried current (n)
            if injury_status == 0: #Not injuried (n-1)
                m=sorted_data.iloc[i,8]
                if m != 0:   
                    sorted_data.iloc[i,8]=sorted_data.iloc[i-1,8]*(1-gamma)+sorted_data.iloc[i,8]*gamma  #Minutes
                    sorted_data.iloc[i,9]=sorted_data.iloc[i-1,9]*(1-gamma)+sorted_data.iloc[i,9]/m *gamma #Points
                    sorted_data.iloc[i,11]=sorted_data.iloc[i-1,11]*(1-gamma)+sorted_data.iloc[i,11]/m *gamma #Rebounds
                    sorted_data.iloc[i,12]=sorted_data.iloc[i-1,12]*(1-gamma)+sorted_data.iloc[i,12]/m *gamma #Steals
                    sorted_data.iloc[i,5]=sorted_data.iloc[i-1,5]*(1-gamma)+sorted_data.iloc[i,5]/m *gamma #Block
                else:
                    sorted_data.iloc[i,8]=sorted_data.iloc[i-1,8]*(1-gamma) #Minutes
                    sorted_data.iloc[i,9]=sorted_data.iloc[i-1,9]*(1-gamma) #Points
                    sorted_data.iloc[i,11]=sorted_data.iloc[i-1,11]*(1-gamma) #Rebounds
                    sorted_data.iloc[i,12]=sorted_data.iloc[i-1,12]*(1-gamma)#Steals
                    sorted_data.iloc[i,5]=sorted_data.iloc[i-1,5]*(1-gamma) #Block
                    
            elif injury_status == 1: #Injuried last few games (n-1)
                injury_status = 0
                injury_history = sorted_data.iloc[start_index:injury_index,13]
                if np.where(injury_history==False)[0].size != 0:
                    index_last_healthy = np.where(injury_history==False)[0][-1]
                    m=sorted_data.iloc[i,8]
                    if m != 0:   
                        sorted_data.iloc[i,8]=sorted_data.iloc[index_last_healthy,8]*(1-gamma)+sorted_data.iloc[i,8]*gamma  #Minutes
                        sorted_data.iloc[i,9]=sorted_data.iloc[index_last_healthy,9]*(1-gamma)+sorted_data.iloc[i,9]/m *gamma #Points
                        sorted_data.iloc[i,11]=sorted_data.iloc[index_last_healthy,11]*(1-gamma)+sorted_data.iloc[i,11]/m *gamma #Rebounds
                        sorted_data.iloc[i,12]=sorted_data.iloc[index_last_healthy,12]*(1-gamma)+sorted_data.iloc[i,12]/m *gamma #Steals
                        sorted_data.iloc[i,5]=sorted_data.iloc[index_last_healthy,5]*(1-gamma)+sorted_data.iloc[i,5]/m *gamma #Block
                    else:
                        sorted_data.iloc[i,8]=sorted_data.iloc[index_last_healthy,8]*(1-gamma) #Minutes
                        sorted_data.iloc[i,9]=sorted_data.iloc[index_last_healthy,9]*(1-gamma) #Points
                        sorted_data.iloc[i,11]=sorted_data.iloc[index_last_healthy,11]*(1-gamma) #Rebounds
                        sorted_data.iloc[i,12]=sorted_data.iloc[index_last_healthy,12]*(1-gamma)#Steals
                        sorted_data.iloc[i,5]=sorted_data.iloc[index_last_healthy,5]*(1-gamma) #Block
            
            elif injury_status == 2: # Injuried since first game (This will be the new "first" game)
                injury_status = 0
                start_index = i
                m=sorted_data.iloc[i,8]
                if m != 0:
                    sorted_data.iloc[i,8]=sorted_data.iloc[i,8]  #Minutes
                    sorted_data.iloc[i,9]=sorted_data.iloc[i,9]/m #Points
                    sorted_data.iloc[i,11]=sorted_data.iloc[i,11]/m #Rebounds
                    sorted_data.iloc[i,12]=sorted_data.iloc[i,12]/m #Steals
                    sorted_data.iloc[i,5]=sorted_data.iloc[i,5]/m #Block
                
        else: #If Injuried (n)
            injury_status = 1
            injury_index = i
           
    #New player initalize
    else:
        if sorted_data.iloc[i,13] == False: #Not injuried first game
            start_index = i
            m=sorted_data.iloc[i,8]
            if m != 0:
                sorted_data.iloc[i,8]=sorted_data.iloc[i,8]  #Minutes
                sorted_data.iloc[i,9]=sorted_data.iloc[i,9]/m #Points
                sorted_data.iloc[i,11]=sorted_data.iloc[i,11]/m #Rebounds
                sorted_data.iloc[i,12]=sorted_data.iloc[i,12]/m #Steals
                sorted_data.iloc[i,5]=sorted_data.iloc[i,5]/m #Block
        else: #Injuried first game 
            injury_status = 2

sorted_data.to_pickle("preprocessed.pkl")
```
